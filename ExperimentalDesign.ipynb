## 概要
---

本notebookでは，実験計画法の概要について調査し，その上で基本的な三原則(局所最適化，反復，無作為化)についても触れる．

## 実験計画法
---

実験計画法はR.A.Fisherによって提案された統計学の応用分野である．ある要因が結果にどのくらい影響を与えるのか分析する際に，大きく分けて以下の2つの観点から方法論を提供する．

- どのように実験を行えば，結果の要因を適切に分析できるか
- どのように実験を行えば，効率的に要因を分析できるか

ここでは，前者の例として基本的な三原則(局所最適化，反復，無作為化)について触れ，後者の例として直行表について触れる．

## 基本的な三原則
---

実験計画法における基本的な三原則は，幅広い分野で適切にデータを取得するための基本原理を提供してくれる．

### 局所最適化・局所管理化

実験において調べる要因(因子)以外の要素をできるだけ統一するという原則．例えば，施肥量を因子とした作物の収穫量を調べるとき，施肥量以外の因子(気温，栽培場所，給水量など)はできるだけ揃えるようにする．

### 反復

データを取得した際に偶然起こるバラツキを軽減するために，複数回データを採取すること．

### 無作為化

局所最適化や反復を行っても取り除けない因子による影響を取り除くため，データ採取の際にランダム性を持たせて偏りを軽減すること．例えば，実験手順による影響を取り除くため，入れ替え可能な手順については毎度異なる手順で実験を行うことが挙げられる．

## 直行表
---
要因と考えられているもの(因子)が結果に影響を与えるかどうかを調べる要因分析であるが，因子の数や因子のとりうる内容(水準)が増えれば増えるほど，採らなければならない実験のデータは指数的に増えていってしまう．

例えば，「起床時刻が遅くなる」という結果に対して「ゲームが趣味である」「お酒を飲む」「夜勤がある」という3つの因子の影響を調べるとする．それぞれの因子は「YES」「NO」の2つの値を取りうるので，水準は2である．したがって，全通りのパターンを素朴に考えると$2^3=8$通りのデータを採取する必要がある．この例でいえばまだ少ない方であるが，仮に4因子3水準の実験をすることになった場合はどうだろうか．$3^4=81$通りものデータを採らなければならなくなる．

このように総当たりでデータ採取の実験計画をするのではなく，直行表を使用することで大幅に実験データの取得回数を削減することができる．直行表には様々なパターンがあるが，今回は$L4(2^3)$直行表を例示する．

|  No   |  ゲーム   |  飲酒   |  夜勤   |
| --- | --- | --- | --- |
|  1   |  No   |  No   |  No   |
|  2   |  No   |  Yes   |  Yes   |
|  3   |  Yes   |  No   |  Yes   |
|  4   |  Yes   |  Yes   |  No   |
|  水準数   |  2   |  2   |  2   |

表から分かるように，すべての因子について2回ずつ水準が現れている．因子の間で交互作用(片方の因子がもう片方の因子に影響を与えること)がなければ，このように半分の4通りの実験で済む．交互作用が存在するだろうと予想できる場合には，大きめの直行表に対して因子として交互作用を組み込むこともある．

この他にも3水準の因子4つまでを扱える$L9(3^4)$直行表なども存在する．
