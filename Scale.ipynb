# 名義尺度，順序尺度，間隔尺度，比例尺度の違いを調べる(2021/01/06)
---

## 名義尺度
---
事例を区別するために用いられる尺度のこと．
例えば，「1:男，2:女, 3:その他」のように割り当てたり，地域に郵便番号を割り当てたりすることは名義尺度である．
そのため，名義尺度同士の四則演算には意味がない．すなわち先述の例では「男を2倍したら女になる」という表現はできない．

ただし頻度は計算できるため，最頻値は意味を持つ．

## 順序尺度
---
その名の通り，順番に意味を持たせた尺度のこと．
徒競走における「一着，二着，三着，...」や震度，バージョン管理のナンバリングもこれにあたることが多い．

名義尺度のように，それぞれの事例を区別できるのに加え，事例間の前後関係が分かる．ただし同様に尺度同士の四則演算には意味がない．例えば先の例を用いると，「徒競走の三着から一着を引いたら，二着になる」という表現に意味はない．

順序関係を持つため，統計量としては最頻値に加えてパーセンタイルを求めることができる．

## 間隔尺度
---
値の間隔にも意味を持たせたのが間隔尺度である．温度や湿度，西暦などがこれにあたる．

順序尺度の特徴に加えて，順序尺度ではできなかった差の計算が意味を持つ．例えば「2022年の22年前は2000年だ」と表現でき，意味を持つ．
統計量としては平均値や相関係数，標準偏差などを追加で求められる.

ただし，比の計算は意味を持たない点に注意する．例えば「20℃は40℃の2倍冷たい」という表現は意味を持たない．また，「0」という値は「存在しない」という意味ではなく，基準点に過ぎない．

## 比例尺度
---
間隔尺度に加えて，比の計算にも意味を持つのが比例尺度である．質量や速さ，円単位などがこれにあたる．比の計算が可能なので「20km/hは10km/hの2倍速い」という表現が意味を持つ．また，「0」という表現は「存在しない」という意味を持つ．例えば「給料0円」は給料が全く存在しないという意味を持つ．

比例尺度はここまで述べたすべての尺度の特徴を持つ．
- 「10gと10gは同じ質量」(名義尺度)
- 「10gは20gよりも軽い」(順序尺度)
- 「10gと20gの差は10gである」(間隔尺度)

間隔尺度で意味を持つ統計量は比例尺度でも意味を持ち，加えて変動係数も意味を持つ．

## まとめると
---
ここまで整理すると分かるように，名義尺度・順序尺度・間隔尺度・比例尺度はこの順で上下関係がある．統計量は下位の尺度で求められるものであれば，上位の尺度でも求められる．例えば名義尺度は最頻値が意味を持つが，これは間隔尺度においても同様に意味を持つ．
