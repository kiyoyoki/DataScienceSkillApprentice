## 指数関数とlog関数の関係
---
### 指数関数
指数関数は以下のような式で与えられる関数である：

\begin{eqnarray}
y = a^x
\end{eqnarray}

このとき$a$は，$a\neq1$かつ$a>0$であり，「底(てい)」と呼ばれる定数である．このような関数を具体的に「$a$を底とする指数関数」と呼ぶ．

また，一般的に「指数関数」と呼ぶ場合には底が$e$である関数を指すことが多い．$e$はネイピア数と呼ばれる無理数である．これを$y=\exp x$と表現することもある．

底$a$の値によって，指数関数のグラフは2種類の形をとりうる．一つは$0<a<1$の場合であり，一つは$a>1$の場合である．以下，実際にmatplotlibを用いてグラフを描画してみる．

```python
import numpy as np
import matplotlib.pyplot as plt

# 0 < a< 1の場合
a_l = 1/2
x = np.arange(-10, 10, 0.1)
y_l = a_l ** x

plt.plot(x, y_l, label='$0<a<1$')

# 1 < aの場合
a_g = 2
y_g = a_g ** x

plt.plot(x, y_g, label='$1<a$')
plt.legend()
```

### 対数関数
対数関数は以下のような式で表される関数である：

\begin{eqnarray}
y = \log_ax
\end{eqnarray}

この式の意味することは，変数$x, y$の間に$x = a^y$という関係があるということである．つまり，指数関数とは逆関数の関係にある．

指数関数の定義と同様に，$a$を「底(てい)」と呼び，1以外の正の定数である．また，対数関数の変数$x$に相当する部分を「真数(しんすう)」と呼び，先に説明した定義より必ず$0<x$となる．

対数のうち，底の値によって特別に名前がつけられている場合がある．$a=e$の場合を自然対数と呼び，$a=10$の場合を常用対数と呼ぶ．また，情報理論をはじめとする分野では，情報を$\{0,1\}$のバイナリ値で表現する背景から，$a=2$を用いることもある．これを二進対数と呼ぶ．

底$a$の値によって，対数関数のグラフも2種類の形をとりうる．例として，以下にmatplotlibを用いたグラフを描画する：

```python
# 底を直接指定するnumpy関数が見つからなかったので，
# 底の変換公式で間接的に値を求める
# 0 < a< 1の場合
x = np.arange(0.1, 10, 0.1)
a_l = 1/2
y_l = np.log(x)/np.log(a_l)

plt.plot(x, y_l, label='$0<a<1$')

# 1 < aの場合
a_g = 2
y_g = np.log(x)/np.log(a_g) 

plt.plot(x, y_g, label='$1<a$')
plt.legend()
```

### 逆関数の関係をグラフからみる
指数関数と対数関数が逆関数の関係にある様子を，グラフを観察することで確認する．ちょうど$y=x$を軸に対称になっている様子がわかる．

```python
# 0 < a < 1のとき
a = 1/2
# 指数関数
x_exp = np.arange(0, 10, 0.1)
y_exp = a ** x_exp

# 対数関数
x_log = np.arange(0.01, 10, 0.01)
y_log = np.log(x_log)/np.log(a)

# y=xの直線も描画しておく
x = np.arange(0, 10, 0.1)
y = x

plt.plot(x_exp, y_exp, label='exp func')
plt.plot(x_log, y_log, label='log func')
plt.plot(x,y)

plt.xlim([0, 4])
plt.ylim([0, 4])
plt.legend()
```

```python
# 1 < aのとき
a = 2
# 指数関数
x_exp = np.arange(0, 10, 0.1)
y_exp = a ** x_exp

# 対数関数
x_log = np.arange(0.01, 10, 0.01)
y_log = np.log(x_log)/np.log(a)

# y=xの直線も描画しておく
x = np.arange(0, 10, 0.1)
y = x

plt.plot(x_exp, y_exp, label='exp func')
plt.plot(x_log, y_log, label='log func')
plt.plot(x,y)

plt.xlim([0, 4])
plt.ylim([0, 4])
plt.legend()
```

## 片対数グラフ
---
片対数グラフは，グラフを描画する際に片方の軸を対数スケールにし，値の増減が激しいデータを可視化しやすくする．対数スケールでは，目盛りが1変わるごとに値は10倍，20倍，...となる(10を底とした場合)．

片対数グラフを用いる例として，2変数のうち片方が大きく変化するデータが得られた場合が挙げられる．ここでは，一方が指数関数的な増加，もう片方が線形的な増加を示すデータを用いる．これらを同時にプロットすると，グラフの推移が「つぶれて」分からなくなってしまう．

```python
# 指数関数と一次関数を同時にプロットしてみる
x = np.arange(0, 10, 0.1)

y = 1/2 * x
y_exp = np.exp(x)

plt.plot(x, y, label='$y=1/2x$')
plt.plot(x, y_exp, label='$y=\exp x$')
plt.legend()
```

ここで，y軸を対数スケールとした片対数グラフにすることで，データの推移が把握しやすくなる．

```python
plt.yscale('log')
plt.plot(x, y, label='$y=1/2x$')
plt.plot(x, y_exp, label='$y=\exp x$')
plt.legend()
```

指数関数が直線で描画される理由は，以下のように式変形するとわかる：

\begin{eqnarray}
y = \exp x
\end{eqnarray}

両辺で対数を取って

\begin{eqnarray}
\log y &=& \log \exp x \\
&=& \log e^x \\
&=& x \log e \\
&=& x
\end{eqnarray}

$\log y=Y$とおくと，$Y=x$となり，片対数グラフにおいては直線になる．

## 両対数グラフ
---
両対数グラフは片対数グラフの派生で，x軸とy軸両方のスケールを対数にしたものである．変数$x, y$双方が大きく変化するようなデータを描画する際に役立つ．

以下，片対数グラフと同じ関数で，両対数グラフを描画してみる：

```python
plt.xscale('log')
plt.yscale('log')
plt.plot(x, y, label='$y=1/2x$')
plt.plot(x, y_exp, label='$y=\exp x$')
plt.legend()
```

両対数グラフの場合は$y=\frac{1}{2}x$のグラフが直線になった．もともと一次関数は直線で表されるが，両対数グラフの場合には冪関数についても直線で表せるという性質がある．

例えば，$y=\frac{1}{2}x^3$のような三次関数も直線で表せる．先の指数関数の式変形と同様に，関数の両辺で対数を取ることで，この性質は理解できる：

\begin{eqnarray}
y = \frac{1}{2}x^3
\end{eqnarray}

両辺の対数を取ると，

\begin{eqnarray}
\log y &=& \log \frac{1}{2}x^3 \\
&=& \frac{1}{2}\log x^3 \\
&=& \frac{3}{2}\log x \\
\end{eqnarray}

ここで$Y = \log y $，$X = \log x$とおくと，

\begin{eqnarray}
Y = \frac{3}{2}X
\end{eqnarray}

よって両対数グラフにおいては，直線で表現されることがわかる．
