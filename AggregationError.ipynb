# ありがちな集計ミスについて調べる(2022/01/21)
---
## 概要
---
タイトルの通り．データの集計を行った際に発生しやすいミス(とその原因)について，探索的に調査した．

## 事例
---
- 分析に必要な特徴量をDBから抽出し忘れている
- DBに格納されている特徴量を誤って変更する
- 代表値算出の際に，参照範囲を間違える
- 二重計上
- 観測値の採り間違え
- 適切なデータクリーニング処理を行わない
- 集計プログラムの記述ミス
- データの入力ミス
- 0とNULLの設定ミス
- SQLクエリの記述ミス(特にテーブル結合)
- 判断の基準値を集計後に決める
- レポート作成時の転記ミス

## 原因
---
- データを逐次的に手作業で入力している
- 分析手法(手順)が複雑になり，属人化している
- 簡易的なツールの使用
- スキル不足
- 引継ぎがずさん
- 確認(テスト)の不足
- 受託者の要件理解不足
- 仕様書・指示書での要求が曖昧
- 工程期間の圧迫
- 未回答箇所の照会
- 矛盾回答に関する処理方針が曖昧
- 複数回答に関する処理方針が曖昧
- 境界値処理の方針が曖昧
- 調査間で項目が変更される
- 追加の作業指示
- 集計システムの改修
- 導出項目が長大
- 集計目的の理解が曖昧
- 集計値導出のもととなるデータ値を明示していない
- 事実と意見(解釈)が明確に分かれていない
- 再現性がない(集計に用いたデータ・手順が分からない)
- 扱うデータにセット商品が含まれる
