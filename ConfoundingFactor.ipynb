# ある変数が他の変数に与える影響(因果効果)を推定したい場合，その双方に影響を与える共変量(交絡因子)の考慮が重要であると理解する(2022/03/03)
---

## 概要
---

タイトルにある通り，本notebookでは因果推論に関することをメモする．

具体的には，ある原因から結果が生まれる際に，その両方に影響を与える交絡因子(Confounding factor)について調べる．

## 交絡因子
---

交絡因子(Confounding factor, Confounder)とは，原因Aと結果Bに対して以下のような性質を持つ因子・変数Cのことである．

1. 原因Aと関連がある
1. 結果Bに影響を与える
1. AとBの中間因子ではない

ここで表している中間因子の状況は，CがAの結果であり，CがBの原因となるような状況になることを指している．

交絡因子を想定し，交絡を考慮した因果推論を行わないと，誤った推論結果を導き出してしまう可能性がある．AとBは本来独立であるのに，相関関係ひいては因果関係があるとみなしてしまう疑似相関もその一種である．

交絡因子が考慮される代表的なものとして，疾病の因果推論が挙げられる．よく用いられる例として，心筋梗塞・脳梗塞等の循環器系疾患がある．心筋梗塞が発症するという結果に対して，原因はコーヒーにあると考えたとする．しかし，コーヒー等のカフェインを含んだ食材については，適量を摂取することで循環器系の疾病発症リスクを抑える効果があることが一般的に知られている．先ほどの推論結果と矛盾した結果である．コーヒーの摂取量と疾病発症確率に相関があった場合には，この一般論は間違った俗説なのだろうか？

実のところ，これらには交絡因子として喫煙の有無が考えられる．喫煙者にはコーヒーの愛飲者が多く，喫煙は循環器系への負担が大きい．かつ，喫煙は中間因子にはない．喫煙の有無を考慮したデータ採取や推論を行わないと，ともすると誤った結論を導いていしまう．

このように，交絡因子の説明例として疾病の因果推論はよく用いられる．

## 交絡因子への対応策
---

交絡因子に適切に対処しないと，誤った因果推論を行ってしまうリスクがあることは先に述べたとおりである．

では，交絡因子の影響を受けずに因果推論を行う方法はあるのだろうか．結論から述べると，完璧な方法はない．根も葉もないようだがこれは当然のことで，現実の事象は複雑に絡み合っているから，交絡因子を全て挙げることはできないし，そもそも知覚すらできない可能性もあるという話である．

ただ，できるだけ交絡因子の影響を考慮する方策はいくつかある．本節では代表的なものを調査した．

### ケースコントロール研究(case-control studies)・マッチング

予め交絡因子が明確な場合に，その交絡因子が同一である標本同士で比較する方法である．例えば先ほどの循環器系疾患の原因を考える場合に，あらかじめ年齢が交絡因子だと分かる場合には，同年齢の標本を用意して，発症の有無で比較対照する．

### コーホート研究(cohort studies)・限定

同一の交絡因子を持つ集団に限定して因果推論を行うという方法である．前節の循環器系疾患の例で考えると，喫煙が交絡因子だと分かっている場合に非喫煙者のみを集めて実験を行う．

コーホートを複数用意する層化(stratification)を用いることもある．コーホート間で違いがあれば，その差が交絡因子であることが考えられる．

ケースコントロール研究もコーホート研究もどちらも，標本を限定するために標本数が少なくなるリスクを抱えている．

### 多変量解析の変数に交絡因子を加える方法

予め交絡因子が判明している場合には，多変量解析の変数として交絡変数を加えてしまうという方法もある．先ほどまでのアプローチとは異なり，交絡変数の影響を取り除くのではなく，交絡変数の影響もコントロール下に置くという方策である．


### 無作為化

母集団が大きく，かつ交絡因子が不明な場合には，標本を十分な量で無作為抽出するのも有効な方法である．既知・未知の交絡因子によって，測定結果が等しく分散することを利用する．
